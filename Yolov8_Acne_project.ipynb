CODE TO CRESIZE THE IMAGES IF NECESSARY

```python
from PIL import Image
import os

input_dir = '/content/drive/MyDrive/amr_unzipped_files/demo/valid/images'  # Replace with your input images directory
output_dir = '/content/drive/MyDrive/amr_unzipped_files/demo/valid/images_resized'  # Replace with your output images directory

os.makedirs(output_dir, exist_ok=True)

def resize_images(input_dir, output_dir, new_size):
    for filename in os.listdir(input_dir):
        if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
            with Image.open(os.path.join(input_dir, filename)) as img:
                img_resized = img.resize(new_size)
                img_resized.save(os.path.join(output_dir, filename))

new_size = (512, 512)  # Replace with your desired size

resize_images(input_dir, output_dir, new_size)

example_image = Image.open(os.path.join(output_dir, os.listdir(output_dir)[0]))
print(f"Resized image size: {example_image.size}")
example_image.show()
```
